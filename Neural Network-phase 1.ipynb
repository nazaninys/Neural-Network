# Dataloader

```python
class Dataloader:
    
    def __init__(self, data, labels, n_classes, batch_size=None, shuffle=False):

        assert len(data)==len(labels)
        self.__n_classes = n_classes
        self.__batch_size = batch_size
        self.__shuffle = shuffle
        self.__data = data
        self.__onehot_labels = self.__onehot(labels, self.__n_classes)
    
    def __onehot(self, labels, n_classes):
        labels = labels.astype(int)
        onehot_vectors = np.zeros((labels.size, n_classes))
        onehot_vectors[np.arange(labels.size),labels] = 1
        return onehot_vectors
    
    def __shuffle_dataset(self):
        shuffler = np.random.permutation(len(self.__data))
        self.__data = self.__data[shuffler]
        self.__onehot_labels = self.__onehot_labels[shuffler]
        
    
    def __iter__(self):
        
        if self.__shuffle:
            self.__shuffle_dataset()
            
        if self.__batch_size==None:
            yield (np.matrix(self.__data), np.matrix(self.__onehot_labels))
            return
            
        for idx in range(0, len(self.__data), self.__batch_size):
            yield (np.matrix(self.__data[idx:idx+self.__batch_size]), 
                   np.matrix(self.__onehot_labels[idx:idx+self.__batch_size]))
```

# Activation Functions

```python
class Identical:
    
    def __init__(self): pass
    
    def __val(self, matrix):
        identical_value = np.matrix(matrix, dtype=float)
        return identical_value

    def derivative(self, matrix):
        temp = np.matrix(matrix, dtype=float)
        identical_derivative = np.matrix(np.full(np.shape(temp), 1.))
        return identical_derivative
    
    def __call__(self, matrix):
        return self.__val(matrix)
    

class Relu:
    
    def __init__(self): pass
    
    def __relu(self, matrix):
        relu_value = np.matrix(matrix, dtype=float)
        relu_value = np.where(relu_value > 0, relu_value, 0)
        return relu_value

    def derivative(self, matrix):
        relu_derivative = np.matrix(matrix, dtype=float)
        relu_derivative = np.where(relu_derivative > 0, relu_derivative, 0)
        relu_derivative = np.where(relu_derivative <= 0, relu_derivative, 1)
        return relu_derivative
    
    def __call__(self, matrix):
        return self.__relu(matrix)

    
class LeakyRelu:
    def __init__(self, negative_slope=0.01):
        self.negative_slope = 0.01
    
    def __val(self, matrix):
        leacky_relu_value = np.matrix(matrix, dtype=float)
        leacky_relu_value = np.where(leacky_relu_value > 0, leacky_relu_value, self.negative_slope * leacky_relu_value)
        return leacky_relu_value

    def derivative(self, matrix):
        leacky_relu_derivative = np.matrix(matrix, dtype=float)
        leacky_relu_derivative = np.where(leacky_relu_derivative <= 0, leacky_relu_derivative, 1)
        leacky_relu_derivative = np.where(leacky_relu_derivative > 0, leacky_relu_derivative, self.negative_slope)
        return leacky_relu_derivative
    
    def __call__(self, matrix):
        return self.__val(matrix)

    
class Sigmoid:
    
    def __init__(self): pass

    def __val(self, matrix):
        sigmoid_value = np.matrix(matrix, dtype = float)
        sigmoid_value = 1 / (1 + np.exp(-1 * sigmoid_value))
        return sigmoid_value

    def derivative(self, matrix):
        sigmoid_derivative = np.matrix(matrix, dtype = float)
        y = 1 / (1 + np.exp(-1 * sigmoid_derivative))
        second = 1 - y
        sigmoid_derivative = np.multiply(y, second)
        return sigmoid_derivative
    
    def __call__(self, matrix):
        return self.__val(matrix)


class Softmax:
    
    def __init__(self): pass

    def __val(self, matrix):
        softmax_value = np.matrix(matrix, dtype = float)
        maxi = np.max(softmax_value, axis = 1)
        softmax_value = softmax_value - maxi
        softmax_value = np.exp(softmax_value) / np.sum(np.exp(softmax_value), axis = 1)
        return softmax_value

    def derivative(self, matrix):
        return softmax_derivative
    
    def __call__(self, matrix):
        return self.__val(matrix)
```

# Loss Function

```python
class CrossEntropy: #(with softmax)
    
    def __init__(self): pass

    def __val(self, true_val, expected_val):
        assert np.shape(true_val)==np.shape(expected_val)
        sfmax = Softmax()(true_val)
        cross_entropy_value = -1 * np.multiply(expected_val, np.log(sfmax + 1e-9))
        cross_entropy_value = np.sum(cross_entropy_value, axis = 1)
        return cross_entropy_value
        
    def derivative(self, true_val, expected_val):
        assert np.shape(true_val)==np.shape(expected_val)
        sfmax = Softmax()(true_val)
        cross_entropy_derivative = sfmax - expected_val
        return cross_entropy_derivative
    
    def __call__(self, true_val, expected_val):
        return self.__val(true_val, expected_val)
```

# Layer

```python
class Layer:

    DEFAULT_LOW, DEFAULT_HIGH, DEFAULT_MEAN, DEFAULT_VAR = 0, 0.05, 0., 1.
  
    def __init__(self, input_size, output_size, 
                 activation=Identical(), initial_weight='uniform', **initializing_parameters):
        
        assert type(initial_weight)==str, 'Undefined activation function!'
        #assert initial_weight in self.__weight_initializer_dict, 'Undefined weight initialization function!'
        
        self.__weight_initializer_dict = {'uniform':self.__uniform_weight, 'normal':self.__normal_weight}
        
        self.__n_neurons = output_size
        weight_initializer = self.__weight_initializer_dict[initial_weight]
        self.__weight = weight_initializer(input_size, self.__n_neurons, **initializing_parameters)
        self.__bias = weight_initializer(1, self.__n_neurons, **initializing_parameters)
        self.__activation = activation
        
        self.__last_input = None
        self.__last_activation_input = None
        self.__last_activation_output = None
        self.__last_activation_derivative = None
        
    def forward(self, layer_input):
        assert np.ndim(layer_input)==2
        assert np.size(self.__weight,0) == np.size(layer_input,1)
        self.__last_input = layer_input
        self.__last_activation_input = np.dot(layer_input, self.__weight) + self.__bias
        self.__last_activation_output = self.__activation(self.__last_activation_input)
        self.__last_activation_derivative = self.__activation.derivative(self.__last_activation_input)
        return self.__last_activation_output
    
    def update_weights(self, backprop_tensor, lr):
        assert np.ndim(backprop_tensor)==2
        assert np.size(backprop_tensor,0) == np.size(self.__last_activation_derivative,0)
        assert np.size(backprop_tensor,1) == self.__n_neurons
        y = np.multiply(backprop_tensor, self.__last_activation_derivative)
        weight = self.__weight
        self.__weight = self.__weight - lr * np.matmul(np.transpose(self.__last_input), y)
        matrix = np.full((1, y.shape[0]), 1)
        self.__bias -= lr * np.matmul(matrix, y)
        backprop_tensor = np.matmul(y, np.transpose(weight))
        
        return backprop_tensor

    def __uniform_weight(self, dim1, dim2, **initializing_parameters):
        low, high = self.DEFAULT_LOW, self.DEFAULT_HIGH
        if 'low' in initializing_parameters.keys(): low = initializing_parameters['low']
        if 'high' in initializing_parameters.keys(): high = initializing_parameters['high']
        weights = np.random.uniform(low, high, (dim1,dim2))
        return weights

    def __normal_weight(self, dim1, dim2, **initializing_parameters):
        mean, var = self.DEFAULT_MEAN, self.DEFAULT_VAR
        if 'mean' in initializing_parameters.keys(): mean = initializing_parameters['mean']
        if 'var' in initializing_parameters.keys(): var = initializing_parameters['var']
        weights = np.random.uniform(mean, var ** (1/2), (dim1, dim2))
        return weights
    
    @property
    def n_neurons(self): return self.__n_neurons
    
    @property
    def weight(self): return self.__weight
    
    @property
    def bias(self): return self.__bias
    
    @property
    def activation(self): return self.__activation
```

# Feed Forward Neural Network

```python
class FeedForwardNN:
    
    def __init__(self, input_shape):
        
        self.__input_shape = input_shape
        self.__output_shape = None
        
        self.__layers_list = []
        
        self.__lr = None
        self.__loss = None

        
    def add_layer(self, n_neurons, activation=Relu(), initial_weight='uniform', **initializing_parameters):
         
        assert type(n_neurons)==int, "Invalid number of neurons for the layer!"
        assert n_neurons>0, "Invalid number of neurons for the layer!"
        
        n_prev_neurons = self.__input_shape if len(self.__layers_list)==0 else self.__layers_list[-1].n_neurons
        new_layer = Layer(n_prev_neurons, n_neurons, activation, initial_weight, **initializing_parameters)
        self.__layers_list.append(new_layer)
        self.__output_shape = self.__layers_list[-1].n_neurons 
      
    
    def set_training_param(self, loss=CrossEntropy(), lr=1e-3):
        assert self.__layers_list, "Uncomplete model!"
        self.__loss = loss
        self.__lr = lr
    
    
    def forward(self, network_input):
        assert type(self.__output_shape) != None, "Model is not compiled!"
        layer_out = self.__layers_list[0].forward(network_input)
        for i in range(1, len(self.__layers_list)):
            layer_out = self.__layers_list[i].forward(layer_out)
        network_output = layer_out      
        return network_output
    
    
    def layer_out(self, trainloader):
        network_output = []
        label = []
        for x_train, y_train in trainloader:
            layer_out = self.__layers_list[0].forward(x_train)
            for i in range(1, len(self.__layers_list) - 1):
                layer_out = self.__layers_list[i].forward(layer_out)
            network_output.append(layer_out) 
            label.append(y_train)
        return network_output, label
    
    
    def fit(self, epochs, trainloader, testloader=None, print_results=True):
        
        assert type(self.__output_shape) != None, "Model is not compiled!"
        assert type(self.__lr) != None and type(self.__loss) != None, "Training paramenters are not set!"

        log = {"train_accuracy":[], "train_loss":[], "test_accuracy":[], "test_loss":[]}
        
        for epoch in range(1, epochs+1):
            
            if print_results: 
                print('Epoch {}:'.format(epoch)) 
                
            average_accuracy, average_loss = self.__train(trainloader)
            log['train_accuracy'].append(average_accuracy)
            log['train_loss'].append(average_loss)
            if print_results:
                print('\tTrain: Average Accuracy: {}\tAverage Loss: {}'.format(average_accuracy, average_loss))
            
            if type(testloader) != type(None):
                average_accuracy, average_loss = self.__test(testloader)
                log['test_accuracy'].append(average_accuracy)
                log['test_loss'].append(average_loss)
                if print_results:
                    print('\tTest: Average Accuracy: {}\tAverage Loss: {}'.format(average_accuracy, average_loss))
                    
        return log
    
    
    def __train(self, trainloader):
        bach_accuracies, batch_losses = [], []
        for x_train, y_train in trainloader:
            batch_accuracy, batch_loss = self.__train_on_batch(x_train, y_train)
            bach_accuracies.append(batch_accuracy)
            batch_losses.append(batch_loss)
        return np.mean(bach_accuracies), np.mean(batch_losses)
    
    
    def __test(self, testloader):
        bach_accuracies, batch_losses = [], []
        for x_test, y_test in testloader:
            batch_accuracy, batch_loss = self.__test_on_batch(x_test, y_test)
            bach_accuracies.append(batch_accuracy)
            batch_losses.append(batch_loss)
        return np.mean(bach_accuracies), np.mean(batch_losses)

    
    def __train_on_batch(self, x_batch, y_batch):
        true_val = self.forward(x_batch)
        entropy = self.__loss(true_val, y_batch)
        batch_average_loss = np.sum(entropy) / len(entropy)
        batch_accuracy = self.__compute_accuracy(true_val, y_batch)
        self.__update_weights(true_val, y_batch)
        return (batch_accuracy, batch_average_loss)
        
        
    def __test_on_batch(self, x_batch, y_batch):
        true_val = self.forward(x_batch)
        entropy = self.__loss(true_val, y_batch)
        batch_average_loss = np.sum(entropy) / len(entropy)
        batch_accuracy = self.__compute_accuracy(true_val, y_batch)
        
        return (batch_accuracy, batch_average_loss)
            
        
    def __get_labels(self, outputs):
        labels = outputs.argmax(axis=1)
        return labels
    
    
    def __compute_accuracy(self, output, expected_output):
        output = self.__get_labels(output)
        expected_output = self.__get_labels(expected_output)
        
        summ = (np.sum(output == expected_output))
        accuracy = (summ / output.size) * 100
        
        return accuracy
    
    
    def __update_weights(self, output, y_train):
        backprop = CrossEntropy().derivative(output, y_train)
        backprop = self.__layers_list[-1].update_weights(backprop, self.__lr)
        for i in range(len(self.__layers_list) - 2, -1, -1):
             backprop = self.__layers_list[i].update_weights(backprop, self.__lr)
            
        
        return
```

In contrast, if its value is too small, it gradually fits to the optimal solution and needs more epochs to get the satisfying solution. It also may lead the model to get stuck.

## Impact of Activation Function

```python
INPUT_SHAPE = 784
LEARNING_RATE = 0.001
EPOCHS = 30
TRAINLOADER = Dataloader(train_data, train_label, 10, batch_size=32, shuffle=True)
TESTLOADER = Dataloader(test_data, test_label, 10, batch_size=32, shuffle=True)

network = FeedForwardNN(INPUT_SHAPE)
network.add_layer(20, input_shape=INPUT_SHAPE, activation=Sigmoid(), weight_initializer='uniform')
network.add_layer(15, activation=Sigmoid(), weight_initializer='uniform')
network.add_layer(10, activation=Identical(), weight_initializer='uniform')
network.set_training_param(loss=CrossEntropy(), lr=LEARNING_RATE)

log = network.fit(EPOCHS, TRAINLOADER, TESTLOADER)
log
```

Activation functions determine the output of a deep learning model, its accuracy, and also the computational efficiency of training a model. They also have impact on network's ability to converge and its converge speed. Some activation functions can normalize the input and change their value between 0 and 1 or -1 and 1. The activation function we use should be computationally efficient because it is implemented on a large number of samples.

As the results of table below shows, Leaky Relu is the best activation function we can use for our model. Sigmoid has sufficient accuracy but it learns so slowly that in the first epochs the accuracy is not acceptable.

Relu is a computationally efficient activation function and it allows the network to converge quickly.  
When the input is negative, the result of activation function is zero which means that some neurons may not be activated. So our network will be sparse. On the contrast, using Sigmoid or Tanh activates all the neurons which makes the activation dense. Disadvantage of Relu is when the input is negative, the gradient becomes zero and we cannot use it for backpropagation and network cannot learn from them. This problem will be solved using Leaky Relu which has a small positive slope in the negative area.

Using Sigmoid activation function ptevents jumps in output values. It also normalizes the input and the values will be between 0 and 1. Disadvantage of this activation function is that for low or high values of input the gradient changes too slowly to be used for further learning. So the network either cannot learn or it will learn too slowly and many epochs are needed.

|-------------|-----------------------|----------------|----------------------|--------------------|
|Relu|89.18|0.29|86.49|0.38|
|Leaky Relu|89.19|0.29|86.52|0.38|
|Sigmoid|83.43|0.47|82.24|0.50|

In our model batch size of 32 leads to a optimal solution. By using batch size of 128 our coputation speed increases and results are calculated faster. However it doesn't have satisfyting accuracy if the number of epochs is less than 15. After training on each batch, weights get updated so when batch size is large, less updates are implemented and therefore we need more epochs for our network to be trained. 


Using batches have some advantages. First, it requires less memory. Second, after training on each batch weights get updated so we can use fewer epochs and our network trains faster. Using fewer examples(having a small batch size) does not have good effect on our network. The reason is that the error gradient will be highly dependent on the training sample and somehow overfitting occurs and the network looses its generality.

Neural network is trained by iterating over batches of samples and updating of weights. To get a better result, we need to use more than one epoch so that the network learns and has a better result. Using many epochs leads to overfitting. It happens when the network has a good performance on train data but its functionality is poor on test data. In this case, it has learned the noises in the train data and cannot perform acceptably on new data. weights and biases get tuned to the train data. So we should monitor test accuracy and when it decreases we have to stop the learning.
